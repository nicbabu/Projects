```python
data['Previous Rank']= data['Previous Rank'].fillna('Unranked')
data
data.columns
```

```python
filtered_data = data[(data['Country/Territory'] == 'United States of America') & (data['Region'] == 'Americas')]
print(filtered_data['Institution Name'])
filtered_data.shape
```

```python
data['Status']= data['Status'].fillna('Unknown')
data['SUS SCORE']=data['SUS SCORE'].ffill()
data
```

```python
data =data. drop(['Size', 'AR RANK','ER RANK', 'FSR SCORE', 'FSR RANK', 'CPF RANK',
       'IFR SCORE', 'IFR RANK', 'ISR SCORE', 'ISR RANK', 'ISD SCORE',
       'ISD RANK', 'IRN SCORE', 'IRN RANK', 'EO SCORE', 'EO RANK',
       'SUS RANK'], axis=1, errors='ignore') # this is less likely important
data
```

```python
#finding the first top 10 university to check its previous rank
import matplotlib.pyplot as plt
import numpy as np
data['2026 Rank(cleaned)']= data['2026 Rank'].astype(str).str.extract(r'(\d+)').astype(float) # it is used to clean the rank columns as it converts to numeric by extracting the first number
data['Previous Rank(cleaned)']= data['Previous Rank'].astype(str).str.extract(r'(\d+)').astype(float)  # it is used to clean the rank columns as it converts to numeric by extracting the first number
top_10_universities = data.sort_values('2026 Rank(cleaned)').head(10)
print(top_universities[['Institution Name', '2026 Rank', 'Previous Rank']])
```

```python
#Plot a simple line chart for rank movement
plt.figure(figsize=(10,6))

for i, row in top_10_universities.iterrows():
    plt.plot([0, 1], [row['Previous Clean'], row['2026 Clean']], marker='o')  # line from previous to 2026
    plt.text(0, row['Previous Clean'], row['Institution Name'], ha='right', fontsize=8)  # label on left
plt.xticks([0, 1], ['Previous Rank', '2026 Rank'])
plt.title('Top 10 Universities: Rank Change')
plt.gca().invert_yaxis()  
plt.grid(True)
plt.show()
```

```python
# plotting top 10 2026 rank university
top10 = data.sort_values('2026 Clean').head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_universities['Institution Name'], top10['2026 Clean'], color='Orange')
plt.xlabel('2026 Rank')
plt.title('Top 10 Universities by 2026 Rank')
plt.gca().invert_yaxis()  #It makes rank 1 at the top
plt.show()
```

```python
top_10_universities = data.sort_values('2026 Rank(cleaned)').head(10)
scores = top_10_universities[['Institution Name', 'AR SCORE', 'CPF SCORE']]
print(scores)

correlation = scores[['AR SCORE', 'CPF SCORE']].corr()
print("Correlation between AR SCORE and CPF SCORE:")
print(correlation)
```

```python
# To plot the scatter plot of the given 
plt.figure(figsize=(8, 6))
plt.scatter(scores['AR SCORE'], scores['CPF SCORE'], color='green')

for i, row in scores.iterrows():
    plt.text(row['AR SCORE'] + 0.5, row['CPF SCORE'], row['Institution Name'], fontsize=8)

plt.xlabel('AR SCORE', color='Blue')
plt.ylabel('CPF SCORE', color='Blue')
plt.title('AR SCORE vs CPF SCORE (Top 10 Universities)', color="Red")
plt.grid(True)
plt.show()
```

```python
research_strength = top_10_universities[['Institution Name','2026 Rank(cleaned)', 'CPF SCORE','Research','Focus']] #to find the which university is more focused on research, high rank university also has high CPF or reseach values, also reach shows the overall contribution to the ranking.
print(research_strength.sort_values('CPF SCORE', ascending= False))
```
